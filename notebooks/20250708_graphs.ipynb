```python
import networkx as nx
import matplotlib.pyplot as plt
```

## Grafo no dirigido

```python
G = nx.Graph()

# Add nodes
G.add_nodes_from(['A', 'B', 'C', 'D'])
# Add edges
G.add_edges_from([('A', 'B'), ('B', 'C'), ('C', 'D'), ('D', 'A')])

# Graphic
pos  = nx.circular_layout(G)
plt.figure(figsize=(6,4))
nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=500)
plt.title("Grafo no dirigido")
plt.show()
```

## Grafo dirigido

```python
DG = nx.DiGraph()
# agregar aristas
DG.add_edges_from([('A', 'B'), ('B', 'C'), ('C', 'D'), ('D', 'A')])
# Graphic
pos  = nx.circular_layout(DG)
plt.figure(figsize=(6,4))
nx.draw(DG, pos, with_labels=True, node_color='lightblue', node_size=500)
plt.title("Grafo dirigido")
plt.show()
```

```python
DGP = nx.DiGraph()
# agregar aristas
DGP.add_edge('A', 'B', weight=4)
DGP.add_edge('B', 'C', weight=2)
DGP.add_edge('C', 'D', weight=5)
DGP.add_edge('D', 'A', weight=1)
# Graphic
pos  = nx.spring_layout(DGP, k=0.3)
plt.figure(figsize=(6,4))

nx.draw(DGP, pos, with_labels=True, node_color='lightblue', node_size=500)
labels = nx.get_edge_attributes(DGP, "weight")
nx.draw_networkx_edge_labels(DGP, pos, edge_labels=labels)
plt.title("Grafo dirigido y ponderado")
plt.show()
```
