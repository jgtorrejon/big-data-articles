## NLP Pipeline Demo: 5 pasos esenciales

Corpus real: “Emma” de Jane Austen (Project Gutenberg vía NLTK)

```python
# 1. Recolección de corpus
import nltk, textwrap, re

nltk.download("gutenberg", quiet=True)
nltk.download("punkt", quiet=True)
nltk.download('punkt_tab')

from nltk.corpus import gutenberg
raw_text = gutenberg.raw("austen-emma.txt")
print(f"Carácteres en el corpus: {len(raw_text):,}")
print("\nMuestra:\n", textwrap.shorten(raw_text, 250, placeholder="…"))
```

```python
# 2. Limpieza / normalización
# repaso expresiones regulares
```

```python
clean_text = re.sub(r"[^A-Za-z\\s]", " ", raw_text)  # quitar signos y números
clean_text = re.sub(r"\\s+", " ", clean_text)        # colapsar espacios
clean_text = clean_text.lower().strip()
print("\nTexto limpio (primeros 200 c):\n", clean_text[:400], "…")
```

```python
# 3. Tokenizacion
from nltk.tokenize import sent_tokenize, word_tokenize
sentences = sent_tokenize(clean_text)
tokens_per_sentence = [word_tokenize(s) for s in sentences]

print(f"\nTotal de oraciones: {len(sentences):,}")
print("Tokens de la 1.ª oración:", tokens_per_sentence[0][:20])
```

```python
# 4. Stemming vs. Lematización
from nltk.stem import PorterStemmer
stemmer = PorterStemmer()
stemmed = [stemmer.stem(t) for t in tokens_per_sentence[0][:20]]
print("\nPorter stemming:", stemmed)
```

```python
import spacy, sys, subprocess
try:
    nlp = spacy.load("en_core_web_sm")
except OSError:                               # descargar modelo si falta
    subprocess.run([sys.executable, "-m", "spacy", "download", "en_core_web_sm"], check=True)
    nlp = spacy.load("en_core_web_sm")

doc = nlp(" ".join(tokens_per_sentence[0][:20]))
lemmas = [tok.lemma_ for tok in doc]
print("Lemmatización (spaCy):", lemmas)
```

```python
# 5. POS‑tagging & Parsing
print("\nPOS tags y dependencias (primeros 15 tokens):")
for tok in doc[:15]:
    print(f"{tok.text:<12} POS={tok.pos_:<6} DEP={tok.dep_:<10} HEAD={tok.head.text}")
```

```python
# 6. Visualización – Word Cloud
from wordcloud import WordCloud
import matplotlib.pyplot as plt

wc = WordCloud(width=800, height=400, background_color="white",
               max_words=100, collocations=False).generate(clean_text)

plt.figure(figsize=(10, 5))
plt.imshow(wc, interpolation="bilinear")
plt.axis("off")
plt.title("Nube de palabras de 'Emma' – Jane Austen", fontsize=14)
plt.tight_layout()
plt.show()
```
