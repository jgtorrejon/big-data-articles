Por ejemplo analicemos la relación entre Gasto de Transporte y Distancia desde la Residencia al Trabajo

```python
base_gt_drt = base[['Transportation expense', 'Distance from Residence to Work']]
base_gt_drt_r = base_gt_drt.sample(100) #Solo para tener 100 puntos
```

```python
base_gt_drt_r.plot(kind='scatter', x= 'Transportation expense', y= 'Distance from Residence to Work')
```

```python
x = base_gt_drt_r['Transportation expense']
y = base_gt_drt_r['Distance from Residence to Work']
fig = plt.figure()
plt.plot(x, y, 'x', color='g');
```

```python
fig = plt.figure()
plt.plot(x, y, 'o', #color='red',
        markersize=15, 
                        markerfacecolor='white',
                        markeredgecolor='green',
                        markeredgewidth=4)
```

```python
%matplotlib inline
x = base_gt_drt['Transportation expense']
y = base_gt_drt['Distance from Residence to Work']
fit = np.polyfit(x, y, deg=1)

fig = plt.figure()
plt.plot(x, y, '.', color='blue')
plt.xlabel('Gastos de transporte')
plt.ylabel('Distance de la Residencia al trabajo')
plt.plot(x, fit[0] * x + fit[1], color='red') # 
plt.annotate('y = {0:.2f} x + {1:.2f}'.format(fit[0], fit[1]), xy=(300, 40))
plt.show()
'Distancia Trabajo = {0:.3f} * Gasto Transporte + {1:.3f}'.format(fit[0], fit[1]) 
```
